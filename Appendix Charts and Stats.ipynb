## Propaganda categories and ratings

```python
movies_cat = pd.read_csv("./data/movies_cat_df.csv", index_col=0)
movies_cat.head()
```

```python
movies_cat.columns[51:]
```

```python
cat_col = []

for _, row in movies_cat.iterrows():
    for cat in movies_cat.columns[51:]:
        if row[cat] == 1:
            cat_col.append(cat)

cat_col = pd.Series(cat_col)
cat_col.head()
```

```python
movies_cat = movies_cat.assign(cat = cat_col)
movies_cat.head()
```

```python
sns.swarmplot(x=movies_cat.cat, y=movies_cat.ratingValue)
# yeah that didn't turn out so well eh
```

## Effects of genre on rating, with CI

```python
effects = [["Animation**", 0.6637, 0.126],
 ["Suspense", 0.0315, 0.067],
 ["Drama**", 0.5800, 0.057],
 # ["Western", 0.5961, 0.319],
 ["Crime", 0.1476, 0.083],
 ["Disaster", 0.0800, 0.233],
 ["Biography**", 0.6756, 0.171],
 ["Sci-fi", -0.0024, 0.187],
 ["Children", 0.1388, 0.179],
 # ["Musical", -0.2770, 0.390],
 ["Fantasy*", -0.2211, 0.087],
 ["Family**", 0.6565, 0.108],
 ["Comedy**", -0.2698, 0.057],
 # ["音乐", 0.5395, 0.266],
 ["Thriller**", -0.7820, 0.091],
 # ["Sports", -0.0152, 0.230],
 ["History**", 0.3784, 0.140],
 ["Period", -0.0642, 0.092],
 ["Romance**", -0.4068, 0.057],
 ["Adventure*", -0.2277, 0.107],
 ["Horror**", -0.5287, 0.129],
 ["Action**", -0.4003, 0.065],
 ["War", -0.1756, 0.161],
 ["Wuxia", 0.2845, 0.151], ]
```

```python
effects = pd.DataFrame(effects, columns=["genre", "effect", "sd"])
effects.head()
```

```python
ci = effects.sd.apply(lambda x: 1.96 * x)
effects = effects.assign(ci=ci)
effects.head()
```

```python
# https://stackoverflow.com/questions/66895284/seaborn-barplot-with-specified-confidence-intervals
ax = sns.barplot(data=effects, x='genre', y='effect')
ax.errorbar(data=effects, x='genre', y='effect', yerr='ci', ls='', lw=3, color='black')
```

```python
ax.get_figure().savefig("./charts/genre_effect.png", dpi=150)
```
