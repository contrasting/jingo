# Summary Statistics

This notebook seeks to understand the data. It analyses and prints/plots stats like the lowest rated movies, number of movies in each genre, and the average rating of movies for each genre.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# https://stackoverflow.com/a/66513177/10176917
plt.rcParams['font.sans-serif']=['SimHei'] #Show Chinese label
plt.rcParams['axes.unicode_minus']=False   #These two lines need to be set manually
plt.rcParams['figure.figsize'] = (16, 9)
```

```python
movies = pd.read_csv("./data/movies_df.csv", index_col=0)
movies.head()
```

## Distribution of ratings

```python
sns.displot(movies, x="ratingValue", bins=10)
```

```python
# Do movies with high and low rating values also happen to have low rating counts? My hunch is yes
# https://stackoverflow.com/questions/21441259/pandas-groupby-range-of-values
mean_rating_count_by_value = movies.groupby(pd.cut(movies["ratingValue"], np.arange(1, 11, 1))).ratingCount.mean()
mean_rating_count_by_value
```

```python
sns.barplot(x=mean_rating_count_by_value.index, y=mean_rating_count_by_value)
# so clearly there is some sort of relationship between rating count and rating value, though it is not linear
# direction of causation is also not entirely clear - does higher rating lead to more views, or other way?
# could be that people want to see movies with higher ratings, so higher rating => more views.
# Yet as more people see a movie, it becomes more mainstream, dragging the rating down: more views => lower rating
```

## Frequencies of movies in each genre

```python
import json

genres_set = set()

for gs in movies.genre:
    # json requires double quoted strings
    for g in json.loads(gs.replace("'", "\"")):
        genres_set.add(g)

genres_set
```

```python
len(genres_set)
```

```python
# tally movies of each genre

genre_freq = dict.fromkeys(genres_set, 0)

for gs in movies.genre:
    for g in genres_set:
        if g in gs:
            genre_freq[g] += 1

genre_freq = pd.Series(genre_freq)
genre_freq
```

```python
# plot frequencies of movies in each genre
sns.barplot(x=genre_freq.index, y=genre_freq)
```

```python
# what exactly is the sports genre?
movies[movies.genre.str.contains("运动")]
```

```python
# there are Chinese Western movies?
movies[movies.genre.str.contains("西部")]
```

```python
# don't care about these - too low freq or obscure
ignore_genres = {
    "黑色电影",
    "真人秀",
    "戏曲",
    "同性",
    "情色",
    "脱口秀",
    "鬼怪",
}
```

```python
to_ignore = []

for index, row in movies.iterrows():
    for g in ignore_genres:
        if g in row.genre:
            to_ignore.append(index)

len(to_ignore)
```

```python
# which movies contain ignored genres?
movies.iloc[to_ignore]
```

```python
genres_set.difference(ignore_genres)
```

## Number of movies in each year

```python
# plot freq of movies in each year
sns.lineplot(data=movies.year.value_counts())
```

## Average ratings in each year?

```python
year_mean = {year: [] for year in np.arange(2010, 2023)}

for index, row in movies.iterrows():
    year_mean[row.year].append(row.ratingValue)

for year in year_mean:
    year_mean[year] = np.mean(year_mean[year])

year_mean = pd.Series(year_mean)
year_mean
```

```python
sns.lineplot(data=year_mean)
```

```python
sns.swarmplot(x=movies.year, y=movies.ratingValue)
```

## Average ratings of each genre?

```python
genre_mean = {genre: [] for genre in genres_set.difference(ignore_genres)}

for index, row in movies.iterrows():
    for genre in genre_mean:
        if genre in row.genre:
            genre_mean[genre].append(row.ratingValue)

for genre in genre_mean:
    genre_mean[genre] = np.mean(genre_mean[genre])

genre_mean = pd.Series(genre_mean)
genre_mean
```

```python
sns.barplot(x=genre_mean.index, y=genre_mean)
```

Horror is the worst rated genre, followed by thriller, perhaps to little surprise. Western is the highest rated genre, but there are few movies so we must be careful about statistical significance.

## Languages?

```python
language_freq = {}

for index, row in movies.iterrows():
    if type(row.language) is not str:
        continue
    for l in row.language.split("/"):
        ls = l.strip()
        if ls in language_freq:
            language_freq[ls] += 1
        else:
            language_freq[ls] = 1

language_freq = pd.Series(language_freq)
language_freq
```

```python
language_freq = language_freq[language_freq > 5]
language_freq
```

```python
sns.barplot(x=language_freq.index, y=language_freq)
```

```python
# which are the English language movies?
# https://stackoverflow.com/questions/28311655/ignoring-nans-with-str-contains
movies[movies.language.str.contains("英语", na=False)]
# as seen, not all of them are exclusively English, though some like Wonder Woman are
```

```python
# which are the Japanese language movies?
movies[movies.language.str.contains("日语", na=False)]
# again, these movies are hardly "Made in Japan"
```

```python
# which movies have language na?
movies[pd.isnull(movies.language)]
```
